# Function 'loc' to select the rows and columns

```python
education_expenses = df.loc[df['Category'] == 'Education']
groceries_expenses = df.loc[df['Category'] == 'Groceries']
print(f'education_expenses: \n\r {education_expenses}')
print(f'groceries_expenses: \n\r {groceries_expenses}')
```

# Checking expenses above 200

```python
expensive_expenses = df.loc[df['Amount'] >= 200]
expensive_expenses
```

```python
expensive_expenses = df.loc[(df['Amount'] >= 200) & (df['Category'] == 'Groceries')]
expensive_expenses
```

```python
expensive_expenses = df.loc[(df['Amount'] >= 200) & (df['Date']).str.contains('2024')]
expensive_expenses
```

```python
expensive_expenses = education_expenses.loc[(df['Amount'] >= 100) & (df['Date']).str.contains('2024')]
expensive_expenses
```

# Seaborn library

### Importing libraries

```python
import seaborn as sns
```

# Creating scatter plot for the expenses in 2024

```python
expenses_in_2024 = df.loc[(df['Date']).str.contains('2024')]
expenses_in_2024
```

```python
sns.scatterplot(data=expenses_in_2024, x='Date', y='Amount')
```

```python
sns.scatterplot(data=expenses_in_2024, x='Date', y='Amount', hue='Category')
```

```python
expenses_in_2024_months = [vers[5:7] for vers in expenses_in_2024['Date']]
expenses_in_2024_months
```

```python
expenses_in_2024.insert(0, 'Month', value=expenses_in_2024_months)
```

```python
expenses_in_2024_for_education = expenses_in_2024.loc[expenses_in_2024['Category'] == 'Education']
expenses_in_2024_for_education
```

```python
sum_of_expenses_in_2024_for_education = expenses_in_2024_for_education.groupby('Month', as_index=False).sum()
sum_of_expenses_in_2024_for_education
```

```python
sum_of_expenses_in_2024_for_education = sum_of_expenses_in_2024_for_education[['Month', 'Amount']]
sum_of_expenses_in_2024_for_education
```

```python
sns.barplot(data=sum_of_expenses_in_2024_for_education, x='Month', y='Amount')
```

```python
expenses_in_2024 = expenses_in_2024[['Month', 'Category', 'Amount']]
expenses_in_2024
```

```python
plot = sns.catplot(data=expenses_in_2024, x='Category', y='Amount', hue='Category', kind='violin')
plot.set_xticklabels(rotation=90) 
```

```python
plot = sns.displot(data=expenses_in_2024, x='Category')
plot.set_xticklabels(rotation=60)
```
